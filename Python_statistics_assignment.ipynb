1. Difference between descriptive and inferential statistics

Descriptive statistics summarize and describe data using measures like mean, median, charts, and tables. Inferential statistics use sample data to make predictions or conclusions about a population, such as hypothesis testing and confidence intervals.

2. Sampling in statistics

Sampling is the process of selecting a subset of a population. Random sampling gives every individual an equal chance, while stratified sampling divides the population into groups and samples from each group.

3. Mean, median, and mode

Mean is the average, median is the middle value, and mode is the most frequent value. They help understand the center of data.

4. Skewness and kurtosis

Skewness measures data asymmetry. Positive skew means a long right tail. Kurtosis measures data peakedness.

5. Mean, median, and mode using Python

```python
import statistics

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
print("Mean:", statistics.mean(numbers))
print("Median:", statistics.median(numbers))
print("Mode:", statistics.mode(numbers))
```

6. Covariance and correlation

```python
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

cov = np.cov(list_x, list_y)[0][1]
corr = np.corrcoef(list_x, list_y)[0][1]

print("Covariance:", cov)
print("Correlation:", corr)
```
