### Question 1: What is a Decision Tree, and how does it work in the context of classification? [cite: 9]

**Answer:**
A Decision Tree is a supervised machine learning algorithm used for both classification and regression tasks. In the context of classification, it works by splitting the dataset into smaller, more homogeneous subsets based on the features that best separate the classes. It builds a tree-like structure where each internal node represents a "test" or condition on a specific feature, each branch represents the outcome of that test, and each leaf node represents a final class label. The model makes predictions by traversing the tree from the root to a leaf node based on the feature values of the input data.



### Question 2: Explain the concepts of Gini Impurity and Entropy as impurity measures. [cite: 11] How do they impact the splits in a Decision Tree? [cite: 12]

**Answer:**
**Gini Impurity** and **Entropy** are mathematical metrics used to evaluate the quality of a split in a dataset.

* **Gini Impurity:** Measures the probability of incorrectly classifying a randomly chosen element if it was randomly labeled according to the distribution of labels in the node. A Gini score of 0 indicates perfect purity (all elements belong to one class).
    $$Gini = 1 - \sum_{i=1}^{c} p_i^2$$
* **Entropy:** Originating from information theory, it measures the level of uncertainty or disorder in a node. A completely pure node has an entropy of 0.
    $$Entropy = -\sum_{i=1}^{c} p_i \log_2(p_i)$$

**Impact on Splits:** [cite: 12] When building a Decision Tree, the algorithm evaluates potential splits across all features and selects the one that results in the lowest Gini Impurity or Entropy (or equivalently, the highest Information Gain) for the resulting child nodes. By continuously minimizing these impurity measures, the tree organizes the data into the most distinct classes possible.

### Question 3: What is the difference between Pre-Pruning and Post-Pruning in Decision Trees? Give one practical advantage of using each. [cite: 17]

**Answer:**
* **Pre-Pruning (Early Stopping):** The process of halting the growth of the decision tree prematurely before it perfectly classifies the training data. This is done by setting hyperparameters like `max_depth`, `min_samples_split`, or `min_samples_leaf`.
    * *Advantage:* It is computationally faster because it prevents the algorithm from building out unnecessary branches.
* **Post-Pruning:** The process of allowing the tree to grow to its full, maximum depth, and then systematically removing or collapsing branches that do not provide significant predictive power (often using techniques like Cost Complexity Pruning).
    * *Advantage:* It is less likely to miss underlying patterns that might only become apparent after a few deeper splits (avoiding the "horizon effect" sometimes seen in pre-pruning).

### Question 4: What is Information Gain in Decision Trees, and why is it important for choosing the best split? [cite: 19]

**Answer:**
**Information Gain** is the measure of the reduction in Entropy (or impurity) achieved after a dataset is split on a specific feature.

It is calculated by subtracting the weighted average entropy of the child nodes from the entropy of the parent node. It is critical for choosing the best split [cite: 19] because the decision tree algorithm inherently acts greedily—at each node, it selects the feature and threshold that maximizes Information Gain, thereby ensuring that each split brings the model closest to perfectly classifying the target variable.

### Question 5: What are some common real-world applications of Decision Trees, and what are their main advantages and limitations? [cite: 21]

**Answer:**
**Applications:** [cite: 21]
* Medical diagnosis (predicting patient illness based on symptoms).
* Financial sector (credit scoring and loan default prediction).
* Customer churn prediction in marketing.

**Advantages:** [cite: 21]
* Highly interpretable and easy to visualize.
* Requires minimal data preprocessing (no need for scaling or normalization).
* Can handle both numerical and categorical data naturally.

**Limitations:** [cite: 21]
* Highly prone to overfitting (creating complex trees that do not generalize well) if not pruned.
* Can be unstable; a small change in the data can result in a completely different tree structure.

### Question 6: Write a Python program to: Load the Iris Dataset [cite: 31] Train a Decision Tree Classifier using the Gini criterion Print the model's accuracy and feature importances [cite: 32]
Dataset Info: Iris Dataset for classification tasks (sklearn.datasets.load_iris() or provided CSV). [cite: 27]

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# 1. Load the Iris Dataset
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=42)

# 2. Train a Decision Tree Classifier using the Gini criterion
clf_gini = DecisionTreeClassifier(criterion='gini', random_state=42)
clf_gini.fit(X_train, y_train)

# 3. Print the model's accuracy and feature importances
y_pred_gini = clf_gini.predict(X_test)
print(f"Accuracy (Gini): {accuracy_score(y_test, y_pred_gini):.4f}")
print("Feature Importances:", clf_gini.feature_importances_)
```

### Question 7: Write a Python program to: Load the Iris Dataset [cite: 37] Train a Decision Tree Classifier with max_depth = 3 and compare its accuracy to a fully-grown tree. [cite: 38]

```python
# Data already loaded in previous cell

# Fully-grown tree
clf_full = DecisionTreeClassifier(random_state=42)
clf_full.fit(X_train, y_train)
acc_full = accuracy_score(y_test, clf_full.predict(X_test))

# Tree with max_depth=3
clf_depth3 = DecisionTreeClassifier(max_depth=3, random_state=42)
clf_depth3.fit(X_train, y_train)
acc_depth3 = accuracy_score(y_test, clf_depth3.predict(X_test))

print(f"Accuracy (Fully-grown tree): {acc_full:.4f}")
print(f"Accuracy (max_depth=3): {acc_depth3:.4f}")
```

### Question 8: Write a Python program to: Load the Boston Housing Dataset [cite: 45] Train a Decision Tree Regressor [cite: 46] Print the Mean Squared Error (MSE) and feature importances [cite: 47]
Dataset Info: Boston Housing Dataset for regression tasks (sklearn.datasets.load_boston() or provided CSV). [cite: 28, 29]

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
import warnings
warnings.filterwarnings("ignore")

# 1. Load the Boston Housing Dataset
# Note: load_boston() is deprecated in newer versions of scikit-learn.
# We handle it dynamically as requested by the assignment.
try:
    from sklearn.datasets import load_boston
    boston = load_boston()
    X_b, y_b = boston.data, boston.target
except ImportError:
    # Fallback method fetching the raw CSV data if load_boston is removed
    data_url = "http://lib.stat.cmu.edu/datasets/boston"
    raw_df = pd.read_csv(data_url, sep="\\s+", skiprows=22, header=None)
    X_b = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
    y_b = raw_df.values[1::2, 2]

X_train_b, X_test_b, y_train_b, y_test_b = train_test_split(X_b, y_b, test_size=0.3, random_state=42)

# 2. Train a Decision Tree Regressor
reg = DecisionTreeRegressor(random_state=42)
reg.fit(X_train_b, y_train_b)

# 3. Print the Mean Squared Error (MSE) and feature importances
y_pred_b = reg.predict(X_test_b)
mse = mean_squared_error(y_test_b, y_pred_b)

print(f"Mean Squared Error (MSE): {mse:.4f}")
print("Feature Importances:", reg.feature_importances_)
```

### Question 10: Imagine you're working as a data scientist for a healthcare company that wants to predict whether a patient has a certain disease. **Answer:**

1.  **Handle the missing values:** [cite: 62]
    * *Numerical Data:* I would impute missing values using the median (if the distribution is skewed) or the mean (if normally distributed). Alternatively, I could use a model-based approach like KNN imputation.
    * *Categorical Data:* I would impute missing values using the mode (most frequent category) or treat "Missing" as its own distinct category so no information is lost.
2.  **Encode the categorical features:** [cite: 64]
    * I would use One-Hot Encoding for nominal variables (like blood type or gender) to prevent the model from assuming an artificial hierarchy.
    * I would use Label Encoding or Ordinal Encoding for ordinal variables (like pain severity: Low, Medium, High).
3.  **Train a Decision Tree model:** [cite: 65]
    * I would split the cleaned data into training (e.g., 80%) and testing (20%) sets.
    * I would instantiate a `DecisionTreeClassifier` and fit it to the training data to establish a baseline model.
4.  **Tune its hyperparameters:** [cite: 66]
    * To avoid overfitting—a common issue in healthcare data—I would use `GridSearchCV` or `RandomizedSearchCV` coupled with k-fold cross-validation.
    * Key parameters to tune would include `max_depth`, `min_samples_split`, and `min_samples_leaf`.
5.  **Evaluate its performance:** [cite: 67]
    * Accuracy alone is insufficient for healthcare datasets, which are often imbalanced. I would evaluate using **Recall** (to minimize False Negatives—failing to identify a sick patient is dangerous), **Precision**, **F1-Score**, and plot an **ROC-AUC** curve.
    *
6.  **Business value in the real-world setting:** [cite: 68, 69]
    * *Early Detection & Outcomes:* Automatically flagging high-risk patients allows for earlier medical interventions, drastically improving patient survival and recovery rates.
