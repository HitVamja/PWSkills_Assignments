## Theory Questions and Answers

1. **What is NumPy, and why is it widely used in Python?**

Answer: NumPy is a Python library for numerical computing. It is widely used because it provides fast operations on arrays and mathematical functions.

2. **How does broadcasting work in NumPy?**

Answer: Broadcasting allows NumPy to perform operations on arrays of different shapes by automatically expanding smaller arrays.

3. **What is a Pandas DataFrame?**

Answer: A DataFrame is a 2D table-like data structure in Pandas with rows and columns.

4. **Explain the use of the groupby() method in Pandas.**

Answer: groupby() is used to split data into groups and apply functions like sum or mean.

5. **Why is Seaborn preferred for statistical visualizations?**

Answer: Seaborn provides attractive and easy-to-use statistical plots with less code.

6. **What are the differences between NumPy arrays and Python lists?**

Answer: NumPy arrays are faster, use less memory, and support vectorized operations.

7. **What is a heatmap, and when should it be used?**

Answer: A heatmap shows data values using colors and is used to visualize correlations.

8. **What does the term vectorized operation mean in NumPy?**

Answer: It means performing operations on entire arrays without using loops.

9. **How does Matplotlib differ from Plotly?**

Answer: Matplotlib creates static plots, while Plotly creates interactive plots.

10. **What is the significance of hierarchical indexing in Pandas?**

Answer: It allows multiple index levels for complex data representation.

11. **What is the role of Seaborn’s pairplot() function?**

Answer: pairplot() shows relationships between multiple variables.

12. **What is the purpose of the describe() function in Pandas?**

Answer: describe() provides summary statistics of numerical data.

13. **Why is handling missing data important in Pandas?**

Answer: Missing data can affect analysis accuracy, so it must be handled properly.

14. **What are the benefits of using Plotly for data visualization?**

Answer: Plotly provides interactive and web-friendly visualizations.

15. **How does NumPy handle multidimensional arrays?**

Answer: NumPy supports multidimensional arrays using ndarrays.

16. **What is the role of Bokeh in data visualization?**

Answer: Bokeh is used to create interactive visualizations for web browsers.

17. **Explain the difference between apply() and map() in Pandas.**

Answer: apply() works on rows or columns, while map() works on Series values.

18. **What are some advanced features of NumPy?**

Answer: Advanced features include broadcasting, linear algebra, and random sampling.

19. **How does Pandas simplify time series analysis?**

Answer: Pandas provides date-time indexing and resampling functions.

20. **What is the role of a pivot table in Pandas?**

Answer: Pivot tables summarize and reorganize data.

21. **Why is NumPy’s array slicing faster than Python’s list slicing?**

Answer: NumPy slicing returns views instead of copies.

22. **What are some common use cases for Seaborn?**

Answer: Seaborn is used for statistical plots like boxplots and heatmaps.

1. Create a 2D NumPy array and calculate the sum of each row.

```python
import numpy as np

arr = np.array(([1,2,3,4,5,6],[1,2,3,4,5,6]))
sum = []

for i in arr:
   total = 0
   total = np.sum(i)
   sum.append(total)

sum
```

3. Create a scatter plot using Matplotlib.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

arr = np.array([
    [1, 6],
    [2, 7],
    [3, 8],
    [4, 9],
    [5, 10]
])

x = arr[:, 0]
y = arr[:, 1]

plt.scatter(x, y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Scatter Plot from 2D Array")
plt.show()

```

4. Calculate a correlation matrix and visualize it using a heatmap.

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Maths': [80, 85, 90, 95, 70],
    'Science': [78, 88, 92, 96, 72],
    'English': [75, 82, 88, 90, 68]
}

df = pd.DataFrame(data)

corr_matrix = df.corr()
corr_matrix




plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

8. Create a line plot with multiple lines using Matplotlib.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

plt.plot(x, y1, label="Line 1")
plt.plot(x, y2, label="Line 2")

plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Multiple Line Plot")
plt.legend()
plt.show()
```

13. Create a 3D scatter plot using Plotly.

```python
import plotly.express as px

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
z = [5, 4, 3, 2, 1]

fig = px.scatter_3d(
    x=x,
    y=y,
    z=z,
    title="3D Scatter Plot"
)

fig.show()
```
