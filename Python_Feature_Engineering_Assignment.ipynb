**1. What is a parameter?**
A parameter is a value that controls the behavior of a machine learning model and is learned from the training data.

**2. What is correlation?**
Correlation is a statistical measure that shows how strongly two variables are related.

**3. What does negative correlation mean?**
Negative correlation means when one variable increases, the other decreases.

**4. Define Machine Learning. What are the main components in Machine Learning?**
Machine Learning is a method where machines learn patterns from data. Main components are data, algorithm, model, and evaluation.

**5. How does loss value help in determining whether the model is good or not?**
Loss value shows the error of a model. Lower loss means better performance.

**6. What are continuous and categorical variables?**
Continuous variables have numeric values (age, salary). Categorical variables have categories (gender, color).

**7. How do we handle categorical variables in Machine Learning?**
We use encoding techniques like Label Encoding and One-Hot Encoding.

**8. What do you mean by training and testing a dataset?**
Training data is used to build the model. Testing data is used to check model performance.

**9. What is sklearn.preprocessing?**
It is a module in scikit-learn used for data scaling and encoding.

**10. What is a Test set?**
A test set is a part of data used to evaluate the trained model.

**11. How do we split data for model fitting in Python?**
We use train_test_split() from sklearn.model_selection.

**12. How do you approach a Machine Learning problem?**
Understand data, perform EDA, preprocess data, build model, evaluate results.

**13. Why do we perform EDA before fitting a model?**
EDA helps understand data patterns, missing values, and outliers.

**14. How can you find correlation between variables in Python?**
Using pandas corr() function.

**15. What is causation? Difference between correlation and causation?**
Causation means one variable causes another. Correlation only shows relation, not cause.

**16. What is an Optimizer?**
An optimizer improves model performance by reducing loss.

**17. Types of optimizers?**
Gradient Descent, SGD, Adam, RMSprop.

**18. What is sklearn.linear_model?**
It contains linear models like Linear Regression and Logistic Regression.

**19. What does model.fit() do?**
It trains the model using training data.

**20. What does model.predict() do?**
It predicts output for new input data.

**21. What is feature scaling?**
Feature scaling brings all values to a similar range.

**22. How does scaling help in ML?**
It improves model accuracy and convergence speed.

**23. How do we perform scaling in Python?**
Using StandardScaler or MinMaxScaler.

**24. Explain data encoding?**
Data encoding converts categorical data into numeric form.

**25. Difference between correlation and causation with example?**
Ice cream sales and drowning are correlated but ice cream does not cause drowning.
