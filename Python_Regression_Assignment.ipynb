**1. What is Simple Linear Regression?**
Simple Linear Regression is a technique used to find the relationship between one independent variable and one dependent variable.

**2. Key assumptions of Simple Linear Regression**
Linearity, independence, homoscedasticity, normal distribution of errors.

**3. What does coefficient m represent?**
The slope m shows how much Y changes when X increases by one unit.

**4. What does intercept c represent?**
The intercept c is the value of Y when X is zero.

**5. How do we calculate slope m?**
Slope is calculated using the least squares formula based on X and Y values.

**6. Purpose of least squares method**
It minimizes the sum of squared differences between actual and predicted values.

**7. Interpretation of R²**
R² shows how much variance in Y is explained by X.

**8. What is Multiple Linear Regression?**
It models the relationship between one dependent variable and multiple independent variables.

**9. Difference between Simple and Multiple Regression**
Simple uses one predictor, Multiple uses more than one.

**10. Key assumptions of Multiple Linear Regression**
Linearity, independence, no multicollinearity, homoscedasticity, normal errors.

**11. What is heteroscedasticity?**
It means unequal variance of errors, which affects model reliability.

**12. How to handle multicollinearity?**
Remove variables, use VIF, or apply regularization.

**13. Encoding categorical variables**
Using One-Hot Encoding or Label Encoding.

**14. Role of interaction terms**
They show combined effect of variables on the target.

**15. Intercept interpretation difference**
In multiple regression, intercept depends on all predictors being zero.

**16. Significance of slope**
Slope determines the direction and strength of prediction.

**17. Role of intercept in context**
It provides baseline value for predictions.

**18. Limitations of R²**
It doesn’t indicate causality or overfitting.

**19. Large standard error meaning**
Coefficient estimate is unreliable.

**20. Identifying heteroscedasticity**
Seen as funnel shape in residual plots.

**21. High R² but low adjusted R²**
Model has irrelevant predictors.

**22. Why scale variables?**
To ensure equal contribution and faster convergence.

**23. What is polynomial regression?**
Regression using polynomial terms.

**24. Difference from linear regression**
Polynomial fits curves, linear fits straight line.

**25. When is polynomial regression used?**
When data shows non-linear pattern.

**26. General equation**
Y = aX² + bX + c (example).

**27. Multiple variables in polynomial regression?**
Yes, it can be extended.

**28. Limitations of polynomial regression**
Overfitting and poor extrapolation.

**29. Model fit methods**
R², cross-validation, error metrics.

**30. Importance of visualization**
Helps understand curve fitting.

**31. Polynomial regression in Python**
Using PolynomialFeatures and LinearRegression.
