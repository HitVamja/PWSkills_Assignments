## Theoretical Questions [cite: 5]

**1. Can we use Bagging for regression problems?** [cite: 9]
Yes, Bagging can be used for regression. It averages the predictions of multiple base regressors (like decision trees) to output a final continuous value.

**2. What is the difference between multiple model training and single model training?** [cite: 10]
Single model training relies on one algorithm to learn patterns and make predictions. Multiple model training (ensemble learning) trains several models and combines their predictions to improve overall stability, accuracy, and generalization.

**3. Explain the concept of feature randomness in Random Forest.** [cite: 11]
In a Random Forest, each time a split is considered in a decision tree, only a random subset of features is evaluated instead of all features. This decorrelates the trees, ensuring they don't all make the exact same splits, which reduces variance.

**4. What is OOB (Out-of-Bag) Score?** [cite: 12]
When using bootstrap sampling (sampling with replacement) in Bagging or Random Forests, about one-third of the data is left out for each tree. The OOB score is the performance metric calculated by passing these "left-out" samples through the trees that did not train on them, serving as a free cross-validation metric.

**5. How can you measure the importance of features in a Random Forest model?** [cite: 13]
Feature importance is typically measured by calculating how much the tree nodes that use a specific feature reduce the overall impurity (like Gini impurity or Mean Squared Error) across all trees in the forest.

**6. Explain the working principle of a Bagging Classifier.** [cite: 14]
A Bagging (Bootstrap Aggregating) Classifier creates multiple subsets of the original training data with replacement. It trains a base classifier (usually a decision tree) on each subset and combines their predictions using majority voting.

**7. How do you evaluate a Bagging Classifier's performance?** [cite: 15]
You evaluate it using standard classification metrics such as Accuracy, Precision, Recall, F1-Score, ROC-AUC, or by using the built-in Out-of-Bag (OOB) score.

**8. How does a Bagging Regressor work?** [cite: 16]
Similar to the classifier, it creates bootstrap samples of the training data, trains a regression model on each, and averages the continuous outputs from all base models to make a final prediction.

**9. What is the main advantage of ensemble techniques?** [cite: 17]
The main advantage is improved predictive performance and robustness. They typically reduce variance (overfitting) and/or bias (underfitting) compared to single models.

**10. What is the main challenge of ensemble methods?** [cite: 18]
They are computationally expensive to train and score, require more memory, and lose the interpretability that a single, simple model (like a single decision tree) provides.

**11. Explain the key idea behind ensemble techniques.** [cite: 19]
The key idea is that the "wisdom of the crowd" is better than an individual. Combining multiple weak or average learners results in a single strong learner.

**12. What is a Random Forest Classifier?** [cite: 20]
It is an ensemble learning method that constructs a multitude of decision trees during training and outputs the class that is the mode of the classes (majority vote) of the individual trees, utilizing both data bootstrapping and feature randomness.

**13. What are the main types of ensemble techniques?** [cite: 21]
The main types are Bagging (e.g., Random Forest), Boosting (e.g., AdaBoost, Gradient Boosting, XGBoost), and Stacking (combining different types of models using a meta-model).

**14. What is ensemble learning in machine learning?** [cite: 22]
It is a paradigm where multiple models (often called "weak learners") are trained to solve the same problem and combined to get better results.

**15. When should we avoid using ensemble methods?** [cite: 23]
Avoid them when model interpretability is strictly required (e.g., for regulatory compliance), when computational resources or latency limits are strict, or when a simple model already perfectly solves the problem.

**16. How does Bagging help in reducing overfitting?** [cite: 24]
By training models on slightly different data subsets and averaging their predictions, Bagging smooths out the noisy, high-variance patterns that individual models might overfit to.

**17. Why is Random Forest better than a single Decision Tree?** [cite: 25]
A single decision tree is highly prone to overfitting the training data. Random Forest prevents this by averaging multiple different trees, leading to better generalization on unseen data.

**18. What is the role of bootstrap sampling in Bagging?** [cite: 26]
Bootstrap sampling creates diverse datasets for each base model by randomly selecting samples with replacement. This diversity is crucial for the ensemble to be effective.

**19. What are some real-world applications of ensemble techniques?** [cite: 27]
Applications include fraud detection, medical diagnosis, risk assessment in finance, recommendation systems, and image classification.

**20. What is the difference between Bagging and Boosting?** [cite: 28]
Bagging trains models independently in parallel and combines them to reduce variance. Boosting trains models sequentially, where each new model tries to correct the errors made by the previous ones, primarily to reduce bias.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification, make_regression, load_breast_cancer
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.ensemble import BaggingClassifier, BaggingRegressor, RandomForestClassifier, RandomForestRegressor, StackingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import accuracy_score, mean_squared_error, roc_auc_score, confusion_matrix, precision_score, recall_score, f1_score, precision_recall_curve
import warnings
warnings.filterwarnings('ignore')

# Common setup for classification
X_c, y_c = make_classification(n_samples=500, n_features=10, random_state=42)
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X_c, y_c, test_size=0.2, random_state=42)

# Common setup for regression
X_r, y_r = make_regression(n_samples=500, n_features=10, random_state=42)
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(X_r, y_r, test_size=0.2, random_state=42)

# 21. Train a Bagging Classifier using Decision Trees and print model accuracy. [cite: 30]
bag_clf = BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=10, random_state=42)
bag_clf.fit(X_train_c, y_train_c)
print(f'21. Bagging Classifier Accuracy: {accuracy_score(y_test_c, bag_clf.predict(X_test_c)):.4f}')

# 22. Train a Bagging Regressor using Decision Trees and evaluate using Mean Squared Error (MSE). [cite: 31]
bag_reg = BaggingRegressor(estimator=DecisionTreeRegressor(), n_estimators=10, random_state=42)
bag_reg.fit(X_train_r, y_train_r)
print(f'22. Bagging Regressor MSE: {mean_squared_error(y_test_r, bag_reg.predict(X_test_r)):.4f}')

# 23. Train a Random Forest Classifier on the Breast Cancer dataset and print feature importance scores. [cite: 32]
data = load_breast_cancer()
rf_cancer = RandomForestClassifier(random_state=42).fit(data.data, data.target)
print(f'23. Breast Cancer Top 3 Feature Importances: {rf_cancer.feature_importances_[:3]}')

# 24. Train a Random Forest Regressor and compare its performance with a single Decision Tree. [cite: 33]
rf_reg = RandomForestRegressor(random_state=42).fit(X_train_r, y_train_r)
dt_reg = DecisionTreeRegressor(random_state=42).fit(X_train_r, y_train_r)
print(f'24. RF Regressor MSE: {mean_squared_error(y_test_r, rf_reg.predict(X_test_r)):.4f} vs DT MSE: {mean_squared_error(y_test_r, dt_reg.predict(X_test_r)):.4f}')

# 25. Compute the Out-of-Bag (OOB) Score for a Random Forest Classifier. [cite: 34]
rf_oob = RandomForestClassifier(oob_score=True, random_state=42).fit(X_train_c, y_train_c)
print(f'25. OOB Score: {rf_oob.oob_score_:.4f}')

# 26. Train a Bagging Classifier using SVM as a base estimator and print accuracy. [cite: 35]
bag_svm = BaggingClassifier(estimator=SVC(), n_estimators=10, random_state=42).fit(X_train_c, y_train_c)
print(f'26. Bagging SVM Accuracy: {accuracy_score(y_test_c, bag_svm.predict(X_test_c)):.4f}')

# 27. Train a Random Forest Classifier with different numbers of trees and compare accuracy. [cite: 36]
for n in [10, 50, 100]:
    rf_n = RandomForestClassifier(n_estimators=n, random_state=42).fit(X_train_c, y_train_c)
    print(f'27. RF ({n} trees) Accuracy: {accuracy_score(y_test_c, rf_n.predict(X_test_c)):.4f}')

# 28. Train a Bagging Classifier using Logistic Regression as a base estimator and print AUC score. [cite: 37]
bag_lr = BaggingClassifier(estimator=LogisticRegression(), n_estimators=10, random_state=42).fit(X_train_c, y_train_c)
y_pred_proba = bag_lr.predict_proba(X_test_c)[:, 1]
print(f'28. Bagging LR AUC: {roc_auc_score(y_test_c, y_pred_proba):.4f}')

# 29. Train a Random Forest Regressor and analyze feature importance scores. [cite: 38]
rf_reg_feat = RandomForestRegressor(random_state=42).fit(X_train_r, y_train_r)
print(f'29. RF Regressor First 5 Feature Importances: {rf_reg_feat.feature_importances_[:5]}')

# 30. Train an ensemble model using both Bagging and Random Forest and compare accuracy. [cite: 39]
print(f'30. Bagging Accuracy: {accuracy_score(y_test_c, bag_clf.predict(X_test_c)):.4f} vs RF Accuracy: {accuracy_score(y_test_c, RandomForestClassifier(random_state=42).fit(X_train_c, y_train_c).predict(X_test_c)):.4f}')

# 31. Train a Random Forest Classifier and tune hyperparameters using GridSearchCV. [cite: 41]
param_grid = {'n_estimators': [10, 50], 'max_depth': [None, 5]}
grid = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=3).fit(X_train_c, y_train_c)
print(f'31. Best params from GridSearch: {grid.best_params_}')

# 32. Train a Bagging Regressor with different numbers of base estimators and compare performance. [cite: 44]
for n in [5, 20]:
    br_n = BaggingRegressor(n_estimators=n, random_state=42).fit(X_train_r, y_train_r)
    print(f'32. Bagging Regressor ({n} est) MSE: {mean_squared_error(y_test_r, br_n.predict(X_test_r)):.4f}')

# 33. Train a Random Forest Classifier and analyze misclassified samples. [cite: 45]
rf_misc = RandomForestClassifier(random_state=42).fit(X_train_c, y_train_c)
predictions = rf_misc.predict(X_test_c)
misclassified = np.where(y_test_c != predictions)[0]
print(f'33. Number of misclassified samples: {len(misclassified)}')

# 34. Train a Bagging Classifier and compare its performance with a single Decision Tree Classifier. [cite: 46]
dt_clf = DecisionTreeClassifier(random_state=42).fit(X_train_c, y_train_c)
print(f'34. Bagging Acc: {accuracy_score(y_test_c, bag_clf.predict(X_test_c)):.4f} vs DT Acc: {accuracy_score(y_test_c, dt_clf.predict(X_test_c)):.4f}')

# 35. Train a Random Forest Classifier and visualize the confusion matrix. [cite: 47]
rf_cm = RandomForestClassifier(random_state=42).fit(X_train_c, y_train_c)
print(f'35. Confusion Matrix:\n{confusion_matrix(y_test_c, rf_cm.predict(X_test_c))}')

# 36. Train a Stacking Classifier using Decision Trees, SVM, and Logistic Regression, and compare accuracy. [cite: 48]
estimators = [('dt', DecisionTreeClassifier()), ('svm', SVC()), ('lr', LogisticRegression())]
stack_clf = StackingClassifier(estimators=estimators, final_estimator=LogisticRegression()).fit(X_train_c, y_train_c)
print(f'36. Stacking Classifier Accuracy: {accuracy_score(y_test_c, stack_clf.predict(X_test_c)):.4f}')

# 37. Train a Random Forest Classifier and print the top 5 most important features. [cite: 49]
rf_top5 = RandomForestClassifier(random_state=42).fit(data.data, data.target)
top_5_idx = np.argsort(rf_top5.feature_importances_)[-5:][::-1]
print(f'37. Top 5 Feature Indices: {top_5_idx}')

# 38. Train a Bagging Classifier and evaluate performance using Precision, Recall, and Fl-score. [cite: 50]
y_pred_bag = bag_clf.predict(X_test_c)
print(f'38. Precision: {precision_score(y_test_c, y_pred_bag):.4f}, Recall: {recall_score(y_test_c, y_pred_bag):.4f}, F1: {f1_score(y_test_c, y_pred_bag):.4f}')

# 39. Train a Random Forest Classifier and analyze the effect of max_depth on accuracy. [cite: 51]
for depth in [3, 10, None]:
    rf_depth = RandomForestClassifier(max_depth=depth, random_state=42).fit(X_train_c, y_train_c)
    print(f'39. RF (depth {depth}) Accuracy: {accuracy_score(y_test_c, rf_depth.predict(X_test_c)):.4f}')

# 40. Train a Bagging Regressor using different base estimators (DecisionTree and KNeighbors) and compare performance. [cite: 52]
bag_dt = BaggingRegressor(estimator=DecisionTreeRegressor(), random_state=42).fit(X_train_r, y_train_r)
bag_knn = BaggingRegressor(estimator=KNeighborsRegressor(), random_state=42).fit(X_train_r, y_train_r)
print(f'40. Bagging DT MSE: {mean_squared_error(y_test_r, bag_dt.predict(X_test_r)):.4f} vs Bagging KNN MSE: {mean_squared_error(y_test_r, bag_knn.predict(X_test_r)):.4f}')

# 41. Train a Random Forest Classifier and evaluate its performance using ROC-AUC Score. [cite: 53]
rf_roc = RandomForestClassifier(random_state=42).fit(X_train_c, y_train_c)
print(f'41. RF ROC-AUC: {roc_auc_score(y_test_c, rf_roc.predict_proba(X_test_c)[:, 1]):.4f}')

# 42. Train a Bagging Classifier and evaluate its performance using cross-validation [cite: 54]
cv_scores = cross_val_score(BaggingClassifier(random_state=42), X_c, y_c, cv=5)
print(f'42. Bagging CV Accuracy: {cv_scores.mean():.4f}')

# 43. Train a Random Forest Classifier and plot the Precision-Recall curve [cite: 55]
precisions, recalls, thresholds = precision_recall_curve(y_test_c, rf_roc.predict_proba(X_test_c)[:, 1])
print(f'43. P-R Curve (first 3 points): Precisions: {precisions[:3]}, Recalls: {recalls[:3]}')

# 44. Train a Stacking Classifier with Random Forest and Logistic Regression and compare accuracy. [cite: 56]
stack_rf_lr = StackingClassifier(estimators=[('rf', RandomForestClassifier())], final_estimator=LogisticRegression()).fit(X_train_c, y_train_c)
print(f'44. Stacking (RF+LR) Accuracy: {accuracy_score(y_test_c, stack_rf_lr.predict(X_test_c)):.4f}')

# 45. Train a Bagging Regressor with different levels of bootstrap samples and compare performance. [cite: 57]
for max_samples in [0.5, 1.0]:
    bag_samples = BaggingRegressor(max_samples=max_samples, random_state=42).fit(X_train_r, y_train_r)
    print(f'45. Bagging (max_samples={max_samples}) MSE: {mean_squared_error(y_test_r, bag_samples.predict(X_test_r)):.4f}')
```
